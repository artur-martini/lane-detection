```python
import cv2 as cv
from glob import glob
import matplotlib.pyplot as plt 
import numpy as np
```

```python
data_path = "./driver_23_30frame/driver_23_30frame/05151640_0419.MP4/" #"\lane-detection\driver_23_30frame\driver_23_30frame\05151640_0419.MP4\"

imgs = glob(data_path+"*.jpg")
labels = glob(data_path+"*.txt")
```

```python
img = cv.cvtColor(cv.imread(imgs[0]), cv.COLOR_BGR2RGB)
plt.figure(figsize=(12, 10))
plt.imshow(img)
plt.show()
```

```python
# extract points from .txt file
f = open(labels[0], "r")
lines = f.readlines()
f.close()

# plot points
img_example = cv.cvtColor(cv.imread(imgs[0]), cv.COLOR_BGR2RGB)

for line in lines:
    coords = line.split(' ')

    points = []
    for i in range(0, len(coords)-1, 2):
        x = int(float(coords[i]))
        y = int(float(coords[i+1]))
        points.append([x, y])

    points = np.array(points, dtype=np.int32)
    points = points.reshape((-1, 1, 2))
    img_example = cv.polylines(img_example, [points], isClosed=False, color=(255, 0, 0), thickness=3)

plt.figure(figsize=(20, 14))
plt.imshow(img_example)
plt.show()
```
